```python
def fun(n_sampl,n_ele):
    mean=[]
    for i in range(0,n_sampl):
        samples=random.sample(range(0,data.shape[0]),n_ele)
        mean.append(data.iloc[samples,0].mean())
    return mean
```

```python
sample_mean={}
sample_mean["100saplme_with_50_ele"]=fun(100,50)
sample_mean["1000saplme_with_50_ele"]=fun(1000,50)
sample_mean["100saplme_with_100_ele"]=fun(100,100)
sample_mean["1000saplme_with_100_ele"]=fun(1000,100)
sample_mean["100saplme_with_1000_ele"]=fun(100,1000)
sample_mean["1000saplme_with_1000_ele"]=fun(1000,1000)
```

```python
plt.figure(figsize=(12,6))
plt.subplots_adjust(hspace=0.4,wspace=0.7)


plt.subplot(2,3,1)
sns.distplot(np.array(sample_mean["100saplme_with_50_ele"]))
plt.title("100saplme_with_50_ele")

plt.subplot(2,3,2)
sns.distplot(np.array(sample_mean["1000saplme_with_50_ele"]))
plt.title("1000saplme_with_50_ele")

plt.subplot(2,3,3)
sns.distplot(np.array(sample_mean["100saplme_with_100_ele"]))
plt.title("100saplme_with_100_ele")

plt.subplot(2,3,4)
sns.distplot(np.array(sample_mean["1000saplme_with_100_ele"]))
plt.title("1000saplme_with_100_ele")

plt.subplot(2,3,5)
sns.distplot(np.array(sample_mean["100saplme_with_1000_ele"]))
plt.title("100saplme_with_1000_ele")

plt.subplot(2,3,6)
sns.distplot(np.array(sample_mean["1000saplme_with_1000_ele"]))
plt.title("1000saplme_with_1000_ele")
```
